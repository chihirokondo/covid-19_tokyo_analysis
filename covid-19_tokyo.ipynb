```python
%matplotlib inline
import datetime
import matplotlib.pyplot as plt
from matplotlib import rcParams
rcParams['font.family'] = 'sans-serif'
rcParams['font.sans-serif'] = ['Hiragino Maru Gothic Pro', 'Yu Gothic', 'Meirio', 'Takao', 'IPAexGothic', 'IPAPGothic', 'Noto Sans CJK JP']
import pandas as pd
```

```python
def read_csv_file():
    # Read CSV file.
    original_data = pd.read_csv('covid-19_tokyo.csv')
    original_data.drop('No', axis=1, inplace=True)
    return original_data
```

```python
def gen_time_series_data(mydata):
    # Extract time series data.
    mydata_cp = mydata['公表_年月日'].value_counts()
    mydata_cp.index = pd.to_datetime(mydata_cp.index)
    mydata_cp.sort_index(inplace=True)

    # Fill missing values.
    dates_df = pd.Series(data=0, index=pd.date_range(start='2020-01-24', end=datetime.date.today(), freq='D'))
    mydata_cp = mydata_cp.combine(dates_df, max, fill_value=0)
    return mydata_cp
```

```python
def date_dep(original_data):
    # Time series.
    date_num_infected = gen_time_series_data(original_data)
    
    # 7-day moving average.
    week_moving_ave = date_num_infected.rolling(window=7, center=True).mean()

    # Create a figure.
    date_num_infected.plot.line(style=['ro'], label='Raw data.')
    week_moving_ave.plot.line(label='7-day moving ave.')
    plt.xlabel("Date")
    plt.ylabel("# of infected peoples in Tokyo (linear scale)")
    plt.legend()
    plt.grid()
    plt.show()
    date_num_infected.plot.line(style=['ro'], label='Raw data.')
    week_moving_ave.plot.line(label='7-day moving ave.')
    #date_num_infected.plot.bar()
    plt.yscale('log')
    plt.xlabel("Date")
    plt.ylabel("# of infected peoples in Tokyo (log scale)")
    plt.legend()
    plt.grid()
    plt.show()
```

```python
def date_dep_fixed_age(original_data):
    # Extract data.
    time_series_0s = original_data[original_data['患者_年代'] == '10歳未満']
    time_series_10s = original_data[original_data['患者_年代'] == '10代']
    time_series_20s = original_data[original_data['患者_年代'] == '20代']
    time_series_30s = original_data[original_data['患者_年代'] == '30代']
    time_series_40s = original_data[original_data['患者_年代'] == '40代']
    time_series_50s = original_data[original_data['患者_年代'] == '50代']
    time_series_60s = original_data[original_data['患者_年代'] == '60代']
    time_series_70s = original_data[original_data['患者_年代'] == '70代']
    time_series_80s = original_data[original_data['患者_年代'] == '80代']
    time_series_90s = original_data[original_data['患者_年代'] == '90代']
    time_series_100s = original_data[original_data['患者_年代'] == '100歳以上']
    
    # Time series.
    time_series_0s = gen_time_series_data(time_series_0s)
    time_series_10s = gen_time_series_data(time_series_10s)
    time_series_20s = gen_time_series_data(time_series_20s)
    time_series_30s = gen_time_series_data(time_series_30s)
    time_series_40s = gen_time_series_data(time_series_40s)
    time_series_50s = gen_time_series_data(time_series_50s)
    time_series_60s = gen_time_series_data(time_series_60s)
    time_series_70s = gen_time_series_data(time_series_70s)
    time_series_80s = gen_time_series_data(time_series_80s)
    time_series_90s = gen_time_series_data(time_series_90s)
    time_series_100s = gen_time_series_data(time_series_100s)

    # 7-day moving average.
    week_moving_ave_0s = time_series_0s.rolling(window=7, center=True).mean()
    week_moving_ave_10s = time_series_10s.rolling(window=7, center=True).mean()
    week_moving_ave_20s = time_series_20s.rolling(window=7, center=True).mean()
    week_moving_ave_30s = time_series_30s.rolling(window=7, center=True).mean()
    week_moving_ave_40s = time_series_40s.rolling(window=7, center=True).mean()
    week_moving_ave_50s = time_series_50s.rolling(window=7, center=True).mean()
    week_moving_ave_60s = time_series_60s.rolling(window=7, center=True).mean()
    week_moving_ave_70s = time_series_70s.rolling(window=7, center=True).mean()
    week_moving_ave_80s = time_series_80s.rolling(window=7, center=True).mean()
    week_moving_ave_90s = time_series_90s.rolling(window=7, center=True).mean()
    week_moving_ave_100s = time_series_100s.rolling(window=7, center=True).mean()
    
    # Create figure.
    #time_series_0s.plot.line(label='0s')
    #time_series_10s.plot.line(label='10s')
    #time_series_20s.plot.line(label='20s')
    #time_series_30s.plot.line(label='30s')
    #time_series_40s.plot.line(label='40s')
    #time_series_50s.plot.line(label='50s')
    #time_series_60s.plot.line(label='60s')
    #time_series_70s.plot.line(label='70s')
    #time_series_80s.plot.line(label='80s')
    #time_series_90s.plot.line(label='90s')
    #time_series_100s.plot.line(label='over 100')
    week_moving_ave_0s.plot.line(label='0s')
    week_moving_ave_10s.plot.line(label='10s')
    week_moving_ave_20s.plot.line(label='20s')
    week_moving_ave_30s.plot.line(label='30s')
    week_moving_ave_40s.plot.line(label='40s')
    week_moving_ave_50s.plot.line(label='50s')
    week_moving_ave_60s.plot.line(label='60s')
    week_moving_ave_70s.plot.line(label='70s')
    week_moving_ave_80s.plot.line(label='80s')
    week_moving_ave_90s.plot.line(label='90s')
    #week_moving_ave_100s.plot.line(label='100s')
    plt.xlabel("Date")
    plt.ylabel("# of infected peoples in Tokyo (7-day moving ave)")
    plt.legend()
    plt.grid()
    plt.show()
```

```python
def age_dep(original_data):
    # Age dependency.
    age_num_infected = original_data['患者_年代'].value_counts()
    age_num_infected.rename(index={'10歳未満':'00代'}, inplace=True)
    age_num_infected.rename(index={'100歳以上':'99代'}, inplace=True)
    age_num_infected.sort_index(inplace=True)
    age_num_infected.rename(index={'00代':'10歳未満'}, inplace=True)
    age_num_infected.rename(index={'99代':'100歳以上'}, inplace=True)

    # Create a figure.
    age_num_infected.plot.bar()
    plt.xlabel("Age")
    plt.ylabel("# of infected peoples in Tokyo")
    plt.show()
```
